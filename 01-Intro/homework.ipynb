```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_parquet("https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet")
df 
```

```python
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])
```

```python
df 
```

```python
df["duration_in_hrs"] = df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]
```

```python
df["duration_in_mins"] = pd.Series([round(x.total_seconds()/60,3) for x in df["duration_in_hrs"]])
df["duration_in_mins"]
```

```python
# dropping the hrs in duration 
del df["duration_in_hrs"]
```

```python
# calculating the standard deviation of the duration in mins feature
std_of_duration_in_mins = np.std(df["duration_in_mins"].values)
std_of_duration_in_mins.item()
```

```python
# getting the count of duration trips which are less than 1 min and greater than 60 mins  which we are considering as outliers 
outliers_count = df.loc[(df["duration_in_mins"] <= 1) | (df["duration_in_mins"] >= 60)]
outliers_count
```

```python
df_not_outliers = df.loc[(df["duration_in_mins"] >= 1) & (df["duration_in_mins"] <= 60)]
df_not_outliers.shape
```

```python
df_selected_features = df_not_outliers[["PULocationID", "DOLocationID"]]
df_selected_features
```

```python
df_selected_features["PULocationID"] = df_selected_features["PULocationID"].astype("str")
df_selected_features["DOLocationID"] = df_selected_features["DOLocationID"].astype("str")
```

```python
df_selected_features.dtypes
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import OneHotEncoder , LabelEncoder
from sklearn.model_selection import train_test_split

main_df_dict = df_selected_features.to_dict(orient="records")

dictvect = DictVectorizer(sparse=True)
X = dictvect.fit_transform(main_df_dict)
X
```

```python
X.shape
```

```python
target = df_not_outliers["duration_in_mins"]
target.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X,target)
preds = model.predict(X)
print(r2_score(preds,target))
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(preds,target))
```
